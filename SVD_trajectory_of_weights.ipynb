```python
#parabaloid
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import FancyArrowPatch

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#creating grid
y = np.linspace(-15.5, 24.5, 100)
x = np.linspace(-15.5, 24.5, 100)
x,y = np.meshgrid(x,y)

#set z values
z = (10.49342831 - (x + y))**2

#label axes
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('cost function')
ax.set_title('(10.49342831 - (w0 + w1))**2')


#plot figure
ax.plot_surface(x,y,z,linewidth=0, antialiased=False, shade = True, alpha = 0.5)

plt.show()
```

```python
# Assuming you have initialized w0, w1, learning_rate, X, Y, and n before this code
w0 = 0
w1 = 7
learning_rate = 0.1
X = [1, 2, 3, 4, 5]
Y = [10.49342831, 11.88210183, 20.79537708, 43.33818787, 29.03169325]
n = len(X)

# Lists to store the values of w0, w1, loss, and gradient vectors during optimization
w0_values = [w0]
w1_values = [w1]
loss_values = [0]
gradient_vectors = []

# Number of iterations
num_iterations = 5  # Show 5 updates

for i in range(num_iterations):
    # Your existing code for stochastic gradient descent
    random_index = np.random.randint(0, n)
    x_i = X[random_index]
    y_i = Y[random_index]

    prediction = w0 + w1 * x_i
    loss = (y_i - prediction)**2

    gradient_w0 = -2 * (y_i - prediction)
    gradient_w1 = -2 * (y_i - prediction) * x_i

    w0 -= learning_rate * gradient_w0
    w1 -= learning_rate * gradient_w1

    # Store the updated values of w0, w1, and loss
    w0_values.append(w0)
    w1_values.append(w1)
    loss_values.append(loss/100)

    # Store the gradient vector for this iteration
    gradient_vectors.append((gradient_w0, gradient_w1))

fig = plt.figure(figsize = (10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the trajectory of weights and gradient vectors in 3D
for i in range(num_iterations):
    ax.plot([w0_values[i], w0_values[i] - learning_rate * gradient_vectors[i][0]],
            [w1_values[i], w1_values[i] - learning_rate * gradient_vectors[i][1]],
            [loss_values[i], loss_values[i]], marker='o', linestyle='-', color='b')

    # Add a text label near the line
    ax.text(w0_values[i],
            w1_values[i],
            loss_values[i], f'({w0_values[i]:.2f}, {w1_values[i]:.2f})', color='black', fontsize=8, ha='left')

    # Add an arrow to represent the gradient vector
    ax.quiver(w0_values[i], w1_values[i], loss_values[i],
              learning_rate * gradient_vectors[i][0], learning_rate * gradient_vectors[i][1], 0,
              color='r', length=10, normalize=True, arrow_length_ratio=0.2)



ax.view_init(elev=20, azim=130)
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('cost function')
# Set equal intervals for x, y, and z axes
ax.set_box_aspect([np.ptp(ax.get_xlim()), np.ptp(ax.get_ylim()), np.ptp(ax.get_zlim())])

plt.show()
```
